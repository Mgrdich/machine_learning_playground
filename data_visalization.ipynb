```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(5, 100, 100)
y = np.linspace(10, 1000, 100)
plt.plot(x,y)
# to actually show the plot
plt.show()
```

```python
plt.plot([1,9,8,6],[6,5,2,1])
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('Current')
plt.ylabel('Voltage')
plt.title("Ohm's law")

# specify the ranges
plt.xlim([20,80])
plt.ylim([200,800])
plt.show()
```

```python
x = np.linspace(0,10,20)
y = x*2
plt.plot(x,y,'r+') # r -> red
plt.xlabel('Current')
plt.ylabel('Voltage')
plt.title("Ohm's law")

plt.show()
```

```python
x = np.linspace(0,5,10)
y = np.linspace(3,6,10)

# plot multiple curves
plt.plot(x,y,'b+',x,y**2,'r+',x,y**3,'g^')
plt.show()
```

```python
x = np.linspace(1,10,100)
y = np.log(x)

# initialize the figure
plt.figure(1)

plt.subplot(121)
plt.title('y = log(x)')
plt.plot(x,y)

plt.subplot(122)
plt.title('y = log(x)**2')
plt.plot(x,y**2)

plt.show()
```

```python
x = np.linspace(1,20,100)
plt.figure(1)

# linear
y1 = x

# cubic
y2 = x**3

# log
y3 = np.log(x)

# exponential
y4 = np.exp(x)


ax1 = plt.subplot(221)
plt.title('linear')
plt.plot(x,y1)

ax2 = plt.subplot(222)
plt.title('cubic')
plt.plot(x,y2)

ax3 = plt.subplot(223)
plt.title('log')
plt.plot(x,y3)

ax4 =plt.subplot(224)
plt.title('exponential')
plt.plot(x,y4)

plt.subplots_adjust(wspace=0.4,hspace=0.4)

plt.show()
```
