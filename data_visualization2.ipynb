```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# styling
sns.set_style('whitegrid')

# data
df = pd.read_csv('data/global_sales_data/market_fact.csv')
```

```python
# kde if probability density function
sns.displot(df['Shipping_Cost'],kind='kde')
plt.show()
```

```python
sns.displot(df['Shipping_Cost'][:200],rug=True)
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(15, 5))
fig.suptitle('Initial Pokemon - 1st Generation')

fig1 = sns.displot(data=df['Sales'],kind='kde')
fig1.set_xlabels('Sales')
fig1.ax = axes[0,0]

fig2 = sns.displot(data=df['Profit'],kind='kde',ax=axes[0,0])
fig2.set_xlabels('Profit')
fig2

fig3 = sns.displot(data=df['Order_Quantity'],kind='kde')
fig3.set_xlabels('Order Quantity')
fig3.ax = axes[0,0]

fig4 = sns.displot(data=df['Shipping_Cost'],kind='kde')
fig4.set_xlabels('Shipping cost')
fig4.ax = axes[0,0]

plt.show()
```
